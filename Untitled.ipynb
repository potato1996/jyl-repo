```python
for i in range(1, 21):
    factor_string = 'F' + '{0:02d}'.format(i)
    filtered_data = ori_data[ori_data[factor_string].notna()] # Need change
    sorted_data = filtered_data.sort_values(factor_string)
    
    number_of_baskets = 10
    basket_returns = np.zeros(number_of_baskets)
    chunk_size = math.floor(sorted_data[factor_string].count() / number_of_baskets)
    for j in range(number_of_baskets):
        start = j * chunk_size
        end = (j + 1) * chunk_size
        basket_returns[j] = sorted_data[start:end]['fwd_returns'].mean()
    
    # Plot
    plt.figure(figsize=(10,5))
    plt.bar(range(number_of_baskets), basket_returns)
    plt.ylabel('Returns')
    plt.xlabel('Basket')
    plt.legend(['Returns of Each Basket'])
    plt.show()
```
